```
import datetime
from bs4 import BeautifulSoup
import urllib.request
```

```
webpage = urllib.request.urlopen('https://search.naver.com/search.naver?&where=news&query=%EC%A0%9C%EC%A3%BC%EB%8F%84&sm=tab_tmr&frm=mr&nso=so:r,p:all,a:all&sort=0')
soup = BeautifulSoup(webpage, 'html.parser')

titlelist = []
textlist = []

for title in soup.find_all('a',"news_tit"):
  titlelist.append(title.get_text())
for text in soup.find_all('a',"api_txt_lines dsc_txt_wrap"):
  textlist.append(text.get_text())


print('오늘의 "제주도" 검색 결과')
for i in range(len(titlelist)):
  print('---------------------------------')
  print('기사 제목 : {}'.format(titlelist[i]))
  print('--> {}'.format(textlist[i]))
```
